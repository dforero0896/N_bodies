```python
# El parámetro de entrada z es una lista de arreglos que contiene el vector de cada uno de los n cuerpos.
def f(z):
    forces_x=np.zeros([n_bodies, n_bodies])
    forces_y=np.zeros([n_bodies, n_bodies])
    for i in range(n_bodies):
        for k in range(n_bodies):
            if i==k:
                forces_x[i, k]=0.
                forces_y[i, k]=0.
            else:
                forces_x[i, k]=(z[i][0]-z[k][0])*(-G*m[i]*m[k]/(np.sqrt((z[i][0]-z[k][0])**2+(z[i][1]-z[k][1])**2)**3))
                forces_y[i, k]=(z[i][1]-z[k][1])*(-G*m[i]*m[k]/(np.sqrt((z[i][0]-z[k][0])**2+(z[i][1]-z[k][1])**2)**3))
    tot_force_x=[]
    tot_force_y=[]
    for i in range(n_bodies):
        tot_force_x.append(sum(forces_x[i, :]))
        tot_force_y.append(sum(forces_y[i, :]))
    tot_force_x=np.array(tot_force_x)
    tot_force_y=np.array(tot_force_y)
    
    
    new_z=[]
    for i in range(n_bodies):
        new_z.append(np.array([z[i][2], z[i][3], tot_force_x[i], tot_force_y[i]]))
    return new_z
                                         
```

```python
def one_Kutta4Step():
    global t
    global Deltat
    global z
    K1=f(z)
    z2=[]
    z3=[]
    z4=[]
    for i in range(n_bodies):
        z2.append(z[i]+Deltat*(1./2.*K1[i]))
    K2=f(z2)
    for i in range(n_bodies):
        z3.append(z[i]+Deltat*(1./2.*K2[i]))
    K3=f(z3)
    for i in range(n_bodies):
        z4.append(z[i]+Deltat*K3[i])
    K4=f(z4)
    for i in range(n_bodies):
        z[i]+= Deltat * (1/6.*K1[i] + 1/3. * K2[i] + 1/3.*K3[i] + 1/6.*K4[i])
    t+= Deltat
    del z2, z3, z4
```

```python
numIter=1000
z=np.copy(r)
#Se da el intervalo de tiempo.
tmax=0.9
t=0.
#Se da el timestep.
Deltat=tmax/numIter
#Se crea el arreglo que guardará la historia
history=[]
for i in range(n_bodies):
    history.append(np.array([[t, z[i][0],z[i][1],z[i][2],z[i][3]]]))

for i in range(numIter):
    one_Kutta4Step()
    for k in range(n_bodies):
        history[k]=np.append(history[k], [[t, z[k][0],z[k][1],z[k][2],z[k][3]]], axis=0)
```
